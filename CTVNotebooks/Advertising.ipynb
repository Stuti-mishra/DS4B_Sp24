```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

```

```python
# Load the data

path = "DataSets/Advertising.csv"
data = pd.read_csv(path)

#df = pd.read_csv(path)[["Metropolitan", "White", "Graduates", "Poverty", "PercentFemale"]].dropna()

```

```python

# Define the model
model = smf.ols('Poverty ~ Graduates', data=data).fit()

# Get the summary as a DataFrame
results_df = pd.DataFrame({
    'Coefficient': model.params,
    'Standard Error': model.bse,
    't Value': model.tvalues,
    'P Value': model.pvalues
})

# Add R-squared and Adjusted R-squared at the end of the table
r_squared = pd.DataFrame({'Coefficient': ['R-squared', model.rsquared, '', ''],
                          'Standard Error': ['Adjusted R-squared', model.rsquared_adj, '', ''],
                          't Value': ['', '', '', ''],
                          'P Value': ['', '', '', '']})

results_df = pd.concat([results_df, r_squared], ignore_index=True)

rounded_results = results_table.iloc[:2, :].round(2)

# Display the formatted table
results_df
```

```python


# Assuming the data is already loaded
X = data['Graduates']  # Predictor variable
y = data['Poverty']    # Response variable

# Adding a constant to the model (intercept)
X = sm.add_constant(X)

# Fitting the linear regression model
model = sm.OLS(y, X).fit()

# Creating the results table
results_table = pd.DataFrame({
    'Coefficients': model.params,
    'Standard Errors': model.bse,
    't Values': model.tvalues,
    'P Values': model.pvalues
})

# Rounding the coefficients, standard errors, t-values, and p-values to 2 decimal places
# For the rows with numeric values
rounded_results = results_table.iloc[:2, :].round(3)

# Adding R-squared and Adjusted R-squared at the bottom of the table
additional_metrics = pd.DataFrame({
    'Coefficients': ['R-squared', 'Adjusted R-squared'],
    'Standard Errors': [model.rsquared, model.rsquared_adj],
    't Values': ['', ''],
    'P Values': ['', '']
})

# Combining the rounded results with the R-squared and Adjusted R-squared rows
final_rounded_table = pd.concat([rounded_results, additional_metrics], ignore_index=True)

# Displaying the final rounded table
print(final_rounded_table)
```
